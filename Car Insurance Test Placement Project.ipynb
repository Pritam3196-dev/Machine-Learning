#### Pros:

##### Reduces dimensionality compared to one-hot encoding.

#### Cons:

```python
def engine_(text):
    
    pattern = (r'(\d+\.\d+)(L)? (\d+-)?(Cylinder)?(Turbo)?')
    
    match = re.search(pattern , str(df['engine_type']))
    
    if match :
        
        return {
            
            'displacement_in_liters' : float(match.group(1)) if match.group(1) else None ,
            
            'cylinders' : match.group(3).replace('-' , '') if match.group(3) else None , 
            
            'Turbo' : 1 if match.group(5) else 0
            
        }
        
    else:
        
        return None
```

```python
engine_info = df['engine_type'].apply(lambda x : engine_(x)).apply(pd.Series)
```

### Detect The Outliers

```python
def outliers_detection(col):
    
    sns.boxplot(df[col])
    plt.show()
    
    
for i in list(df.select_dtypes(exclude = 'object').columns)[:]:
    
    outliers_detection(i)
```

```python
def detecting_outliers(df , n):
    
    for col in df.columns[1:]:
        
        Q1 = df[col].quantile(0.25)
        
        Q2 = df[col].quantile(0.50)
        
        Q3 = df[col].quantile(0.75)
        
        IQR = Q3 - Q1
        
        low_lim = Q1 - (n * IQR)
        
        high_lim = Q3 + (n * IQR)
        
        outliers = (df[(df[col] > high_lim) | (df[col] < low_lim)].shape[0])
        
        print(f'{col} : {outliers}')
        
        print('*'*100)
```

#### ANNOVA TEST

```python
significant_variable = []
for i in df[numeric_columns]:
    
    IV = i 
    DV = "is_claim"  
    
    formula = f"{DV} ~ {IV}"
    
    model = ols(formula, data=df).fit()
    
    anova_table = sm.stats.anova_lm(model, typ=2)  
    
    #print(anova_table)
    
    p_value = anova_table["PR(>F)"][0]
    
    significance_level = 0.05 
    
    if p_value < significance_level:
        
        print(i , " : The variable is significant." , p_value)
        
        significant_variable.append(i)
        
        print('*' * 25)
```
