## Another Approach of uniform Distribution

```python
num_samples = 10000
sample_size = 300

distribution_range = (0,1)
```

## uniform : Is this function which create uniform distribution data and while generateing it took random samples from poppulation

```python
samples = np.random.uniform(distribution_range[0] , distribution_range[1] , (num_samples , sample_size))
```

```python
plt.hist(samples)
plt.show()
```

### This is uniform distribution

### Each sample has 300 data

```python
len(samples[0])
```

### Skewness

```python
pd.DataFrame(samples).skew()
```

#### This is the 10000 samples
#### Generate 10000 samples

```python
len(np.random.uniform(distribution_range[0] , distribution_range[1] , (num_samples , sample_size)))
```

### Mean of samples for to draw normal distribution

```python
samples_mean = np.mean(samples , axis = 1)
```

```python
samples_mean
```

```python
len(np.mean(samples , axis = 1))
```

```python
plt.hist(samples_mean , density = True, bins = 30 , edgecolor = 'black')
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(samples_mean)
plt.show()
```

### Skewness

```python
pd.DataFrame(samples_mean).skew()
```

## Exponential

```python
no_samples  = 1000
sample_size = 50
lambdaa = 2
```

### Exponential : Is this function to generate exponential distribution data and while generateing it took random samples from poppulation

```python
samples = np.random.exponential(scale = 1/lambdaa , size = (no_samples , sample_size))
```

#### Samples are 1000 and every sample has 50 data

```python
len(samples)
```

```python
len(samples[0])
```

```python
plt.hist(samples, bins = 30 , density = True)
plt.show()
```

### This is exponential distribution

#### Skewness before noraml distribution

```python
pd.DataFrame(samples).skew()
```

```python
samples_mean_exp = np.mean(samples , axis = 1)
```

```python
samples_mean_exp
```

```python
len(samples_mean_exp)
```

```python
plt.hist(samples_mean_exp , density = True , bins = 50 , edgecolor = 'black')
plt.show()
```

```python
sns.kdeplot(samples_mean_exp)
plt.show()
```

#### Skewness

```python
pd.DataFrame(samples_mean_exp).skew()
```
