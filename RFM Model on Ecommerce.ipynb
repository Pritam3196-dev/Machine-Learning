#### This is RFM model. means we analyse here Reciency , frequency , monetry of customers.

##### We required customerID because we want to anamyse on customer only this is primary feature and we require Date of Purchase because date of purchase told us how customer recently visited. Then we require InvoiceNo because InvoiceNo tells us how much times customer buy the products actually invoice gets generate when buy something and for everytime that perticular customer buy products that  Invoice generates for that customerID so thats why we can track it by count.(like if any one customer has  2 times invoice generate so we can says he / she bought 2 times )we require price because price tells us about monetry. apart of it we don't have any features for perticular RFM model.

##### We generate RMFscore below

```python
RMFscore = df.groupby('CustomerID').agg({'Date' : lambda x : (latest_date - x.max()).days, 
                                              'InvoiceNo' : lambda x : x.count() ,
                                              'Price' : lambda x : x.sum()})
```

```python
RMFscore
```

```python
RMFscore.rename(columns = {'Date' : 'Recency' , 'InvoiceNo' : 'Frequency' , 'Price': 'Monetry'} , inplace = True)
```

```python
RMFscore
```

#### Insights : Here we make group of those CustomerID who performed multiple transaction so we get unique count of customerID by the groupby function . now we perform aggrigate function on cutomerID in
#### 1. We perform substraction maximum date(most recent date) from latest date. so we get count of days that perticular customer how much days before vsisted
#### 2. Here we calculate count of customers transaction repeatation means any perticular customerID has no of times come in dataset means they perform multiple transaction so we just calculate that count only
#### 3. Here we calculate sum of all repeatations(prices) of that perticulate customers.

#### We segregate values of Recency , Frequency and Monetry in fixed category for our understanding

```python
def Recencyscore(x , p , d):
    
    if x <= d[p][0.25]:
        
        return 1
    
    elif x <= d[p][0.50]:
        
        return 2
    
    elif x <= d[p][0.75]:
        
        return 3
    
    else:
        
        return 4
```

```python
RMFscore.Frequency.describe()
```

```python
def Frequencyscore(x , p  , d):
    
    if x <= d[p][0.25]:
        
        return 4
    
    elif x <= d[p][0.50]:
        
        return 3
    
    elif x <= d[p][0.75]:
        
        return  2
    
    else:
        
        return 1
```

```python
RMFscore.Monetry.describe()
```

```python
def monetryscore(x , p , d):
    
    if x <= d[p][0.25]:
        
        return 4
    
    elif x <= d[p][0.25]:
        
        return 3
    
    elif x <= d[p][0.75]:
        
        return 2
    
    else:
        
        return 1
```

```python
RMFscore
```

#### Here we just pass the data to functions which we create

```python
quantiles
```

```python
RMFscore.columns
```

```python
RMFscore['R'] = RMFscore['Recency'].apply(Recencyscore , args = ('Recency' , quantiles))
RMFscore['F'] = RMFscore['Frequency'].apply(Frequencyscore , args = ('Frequency' , quantiles))
RMFscore['M'] = RMFscore['Monetry'].apply(monetryscore , args = ('Monetry' , quantiles))
```

#### Insights : Here we use apply function for passing values on that perticular columns and create new column. apply func. is the func. its take another func. as an argument. sp we call func by passing its name and pass arguments where first value is the respective columns value like recency , frequency , monetry and 2nd argument is quantile value

```python
RMFscore
```

##### We create single column for sum of this three values means lesser value (sum) model should be very good and if higher value (sum) model should be not good  

```python
RMFscore['RMFvalue'] = RMFscore['R'] + RMFscore['F'] + RMFscore['M']
#RMFscore['RMFvalue'] = RMfscore['R' , 'F' , 'M'].sum(axis = 1)
```

```python
RMFscore
```

#### We make another column for group the this three values in one column

```python
RMFscore['RMFGroup'] = RMFscore.R.map(str) + RMFscore.F.map(str) + RMFscore.M.map(str)
```

##### We segregate customers on basis on basis of  RMFvalue 

```python
loyalty_level = ['Platinum' , 'Diamond' , 'Gold' , 'Silver']
score_cuts = pd.qcut(RMFscore.RMFvalue , q = 4 , labels = loyalty_level)
RMFscore['loyalty_level'] = score_cuts.values
```
